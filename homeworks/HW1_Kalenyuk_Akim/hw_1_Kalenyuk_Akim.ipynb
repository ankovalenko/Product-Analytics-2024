#### Почему

1. **Первый уровень – Основные показатели для DAU**
   - **Retention Rate -> DAU**: Удержание пользователей напрямую влияет на DAU, так как чем выше процент пользователей, возвращающихся каждый день, тем больше стабильный рост DAU.
   - **Stickiness -> DAU**: Служит индикатором того, насколько часто пользователи возвращаются в течение месяца. Чем выше, тем больше будет ежедневная активная аудитория.
   - **CSI -> DAU**: Высокий CSI указывает на лояльность и желание пользователей пользоваться продуктом, что приводит к органическому росту активных пользователей.

2. **Второй уровень – AARRR**.
Каждый показатель первого уровня связан с подгруппами из AARRR-фреймворка, который описывает основные этапы жизненного цикла пользователей:

   - **Acquisition -> Retention Rate**: Чем эффективнее кампания по привлечению, тем выше вероятность, что пользователи вернутся, если им продукт подходит. Тут конечно можно разделить привлечение в рамках продуктового маркетинга (чистый retention) и привлечение новых пользователей, например, в рамках performance маркетинга. 
   - **Activation -> Stickiness**: Если пользователи быстро понимают ценность продукта (скачивают и сразу начинают пользоваться), они с большей вероятностью станут активными и будут чаще возвращаться, повышая Stickiness.
   - **Referral -> Stickiness**: Рекомендации пользователей стимулируют активное использование и привлечение новых пользователей.
   - **Revenue -> Stickiness**: ИМХО, мотивированные пользователи, особенно платные, как правило, остаются вовлеченными и активными, что положительно влияет на DAU и Stickiness. Тут конечно вопрос на обсуждение, потому что деньги как бы должны быть выше в иерархии, но мы вроде как топим за активность, поэтому здесь есть некоторая логика. 

3. **Третий уровень – Метрики вовлеченности и удовлетворенности (HEART)**.
HEART-фреймворк позволяет глубже понять поведение и удовлетворенность пользователей, что может также влиять на метрики AARRR и DAU:

   - **NPS -> Acquisition**: Позитивный опыт и лояльность пользователей увеличивают шансы, что текущие пользователи будут рекомендовать продукт новым пользователям.
   - **Engagement -> Stickiness**: Высокая вовлеченность напрямую повышает Stickiness, так как вовлеченность продукта отражается в регулярном возвращении пользователей.
   - **Adoption -> Activation**: Кратко говоря: хороший онбординг и простой пользовательский путь - это путь к быстрой активации и соответственно к более высокой активности. 
   
#### KPI дашборд

KPI дашборд не должен быть сильно загружен метриками и показывать только динамику основных показателей с нужными срезами и фильтрами для того, чтобы понимать, в каком сегменте и почему возникли те или другие изменения. Поэтому я предлагаю на дэше отображать следующие метрики

1. **DAU**:
   - Главная метрика, которую отслеживает компания.
   - **Срезы и фильтры**: устройства (iOS, Android и т.д.)
   
2. **Retention Rate**:
   - Позволяет выявить, насколько эффективно пользователи остаются активными после первого дня использования.
   - **Срезы и фильтры**: новые пользователи против текущих, когорты по времени регистрации, устройства.

3. **Stickiness**:
   - **Срезы и фильтры**: категории пользователей (новые vs. постоянные), канал привлечения.

5. **K-фактор**:
   - Полезен для понимания органического роста и эффективности фич, которые стимулируют приглашения.
   - **Срезы и фильтры**: по периодам, источникам приглашений.

6. **ARPU**:
   - Удобен для мониторинга доходности и понимания, сколько в среднем каждый пользователь приносит компании. Я не пользуюсь платными функциями ТГ, но вроде как есть куча платных фич, кроме подписок, поэтому метрика имеет смысл. 
   - **Срезы и фильтры**: сегментация по типу пользователей (платящие, бесплатные).

7. **NPS**:
   - Полезен для оценки уровня удовлетворенности и качества продукта.
   - **Срезы и фильтры**: по времени использования.

### Задание 2. 

В поисковую выдачу билетов сайта aviasales была разработана и внедрена скоринговая модель, ранжирующая билеты. Ее цель - помочь пользователю выбрать лучший авиабилет самым быстрым способом, подняв его в топ выдачи.

Пожалуйста, внимательно изучи поисковую выдачу сайта и предложи:

Какими метриками бы ты отслеживал качество работы модели?

Исходя из цели модели, мы должны сделать минимальное кол-во метрик, которые будут отражать основную информацию по процессу, на который влияет модель. Метрики должны отражать, насколько выдача удобна для пользователей, приводит ли к увеличению конверсии и повышению удовлетворенности клиентов, а также насколько эффективно достигается цель быстрого выбора подходящего билета. Метрики должны быть понятны ML-инженерам, аналитикам и бизнесу, а значит мы можем сегментировать их по следующим группам:

1. Метрики ранжирования
2. Пользовательские метрики 
3. Бизнес-метрики

#### 1. **Метрики ранжирования**

   - **Position of Selected Ticket**: показывает, на какой позиции в выдаче пользователи находят и выбирают билет. Чем выше средняя позиция выбранного билета, тем лучше работает модель.
   - **Click-Through Rate**: доля кликов на билет в зависимости от его позиции в выдаче. Высокий CTR на верхних позициях указывает на то, что модель успешно ранжирует более привлекательные для пользователя предложения.
   - **Conversion Rate по позициям**: конверсия с клика в покупку на каждом уровне выдачи. Высокий уровень конверсии на верхних позициях указывает на корректность ранжирования.
   - **DCG и NDCG**: метрики, учитывающие релевантность билетов, где выбранные и приобретенные пользователем билеты должны получать более высокий рейтинг в ранжировании.

#### 2. **Пользовательские метрики**

   - **Average Time to Select a Ticket (Среднее время до выбора билета)**: отражает, насколько быстро пользователи находят подходящий билет. Если после внедрения модели пользователи тратят меньше времени на выбор, это указывает на улучшение ранжирования.
   - **Bounce rate**: Показывает, какой процент людей ушли со страницы выдачи, не произамодействовав ни с каким функционалом. Это можно интерпретировать, как то, что людям на первой странице выпали максимально нерелевантные предложения и они решили не продолжать дальше поиск. Еcли смотреть на всю выборку, метрика может быть нечувствительной - важно смотреть ее сегментировано, где модель может плохо отрабатывать. 

### 3. **Бизнес-метрики**

   - **Conversion Rate на покупку**: процент пользователей, завершивших покупку билета после использования поисковой выдачи. Рост конверсии после внедрения модели указывает на то, что улучшенное ранжирование облегчает пользователям выбор билета.
   - **Revenue per Search**: средний доход от одного поиска. Можно напрямую мерить влияние качества выдачи на деньги. 
   - **CSI**: индекс удовлетворенности, связанный именно с поисковой выдачей и удобством использования интерфейса ранжирования. Повышение указывает на то, что пользователи довольны результатами поиска.
